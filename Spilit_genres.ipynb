```python
import pandas as pd
```

```python
df=pd.read_csv('/content/movies.csv', index_col='movieId')
```

```python
df
```

```python
for index,row in df.iterrows():
    if isinstance(row.genres, str):
        genres = row.genres.split('|')
        for genre in genres:
            if genre not in df.columns:
                df[genre] = 0
            df.loc[index,genre] = 1
df
```
